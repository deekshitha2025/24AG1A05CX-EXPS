### What all manipulations have you done and insights you found?

**Data Manipulations Performed**

To turn the raw data into an analytical structure, the following processing steps were engineered into the pipeline:

Header Whitespace Stripping: Auto-trimmed hidden leading or trailing spaces from column names to eliminate syntax bugs like KeyError.

String Standardization: Standardized all date punctuation marks by transforming inconsistent forward slashes (/) into uniform dashes (-).

Dynamic Datetime Parsing: Handled mixed timestamp formats (parsing rows matching DD-MM-YYYY HH:MM alongside rows capturing seconds DD-MM-YYYY HH:MM:SS) to ensure zero data loss during conversion.

Temporal Slicing: Isolated the absolute hour integers (0 through 23) from the synchronized request timestamps.

Shift Window Binning: Categorized hour sequences into four operational shifts: Morning Rush (4am–10am), Daytime Normal (10am–5pm), Evening Rush (5pm–10pm), and Late Night/Early Hours (10pm–4am).

Fulfillment Masking: Created logical boolean safety masks (Is_Gap) grouping failure modes (Cancelled and No Cars Available) against successful trips.

**The 10 Core Insights Found**

High-Level Supply and Demand

Insight 1: Total Scale of Demand Across the tracking matrix, customers opened their apps and initiated a ride request a total of 6,745 times. This represents the total addressable market size during the recorded period.

Insight 2: Absolute Supply Volume Out of those thousands of requests, drivers successfully picked up passengers and completed the journey exactly 2,831 times. This establishes the true baseline supply capacity of the network.

Insight 3: The Systemic Supply-Demand Gap The platform failed to fulfill 3,914 requests. This means that 58.03% of all booking attempts resulted in failure, leaving more than half of the user base stranded.

**Primary Failure Modes**

Insight 4: Vehicle Scarcity Dominates The absolute largest obstacle to fulfillment is asset scarcity. A total of 2,654 requests failed because "No Cars Available" was triggered, compared to 1,260 requests that failed via driver cancellations.

Insight 5: Geographic Load Balance The market demand is relatively balanced but slightly heavier in town. The platform processed 3,507 requests originating within the City, compared to 3,238 requests originating at the Airport.

**Localized & Shift Patterns**

Insight 6: Localized Airport Friction When isolating the data geographically, the Airport terminal area experienced massive supply friction, resulting in 1,911 unfulfilled ride requests (59.02% failure rate for airport requests).

Insight 7: Localized City Friction The City layout proved just as fragile under localized peak pressure, resulting in 2,003 unfulfilled ride requests (57.11% failure rate for city requests).

Insight 8: The Evening Rush Peak The absolute heaviest traffic load on the entire system occurs during the Evening Rush Hour (5pm to 10pm). This block generates the highest density of incoming consumer requests.

**Root-Cause Structural Bottlenecks**

Insight 9: The Airport Asset Scarcity Bottleneck The operational bottleneck at the Airport is a structural deficit of physical cars. 59.02% of all airport requests end in failure because there are flat-out no cars available at the terminals. Drivers are not actively canceling here; there simply aren't enough vehicles present to handle inbound flight waves.

Insight 10: The City Driver Rejection Bottleneck The operational bottleneck inside the City is driven by driver behavior rather than asset scarcity. During the Morning Rush (4am to 10am), the system suffers a massive wave of driver cancellations. Drivers actively reject airport trips because they want to avoid getting stuck in inbound rush-hour traffic and deadheading back to town empty.

#### Chart - 6

````python
# Chart - 6 visualization code
# Average travel time by day category for completed trips
import pandas as pd
import matplotlib.pyplot as plt

# Convert timestamps
dataset['Request timestamp'] = pd.to_datetime(dataset['Request timestamp'])
dataset['Drop timestamp'] = pd.to_datetime(dataset['Drop timestamp'])

# Create Travel Time in minutes
dataset['Travel Time'] = (
    dataset['Drop timestamp'] - dataset['Request timestamp']
).dt.total_seconds() / 60

# Create Time Category
def get_time_category(hour):
    if 5 <= hour < 10:
        return 'Morning'
    elif 10 <= hour < 17:
        return 'Afternoon'
    elif 17 <= hour < 21:
        return 'Evening'
    elif 21 <= hour < 24:
        return 'Night'
    else:
        return 'Late Night'

dataset['Time Category'] = dataset['Request timestamp'].dt.hour.apply(get_time_category)

# Completed trips only
completed_trips = dataset[dataset['Status'] == 'Trip Completed']

# Average travel time
avg_travel_time = completed_trips.groupby('Time Category')['Travel Time'].mean()

# Order categories
avg_travel_time = avg_travel_time.reindex(
    ['Morning', 'Afternoon', 'Evening', 'Night', 'Late Night']
)

# Plot
plt.figure(figsize=(8,5))
bars = plt.bar(avg_travel_time.index, avg_travel_time.values)

plt.title('Average Travel Time by Time Category')
plt.xlabel('Time Category')
plt.ylabel('Average Travel Time (Minutes)')

for bar in bars:
    plt.text(
        bar.get_x() + bar.get_width()/2,
        bar.get_height(),
        f'{bar.get_height():.1f}',
        ha='center'
    )

plt.show()
````

##### 1. Why did you pick the specific chart?

I used bar chart to find the average range and median.

##### 2. What is/are the insight(s) found from the chart?

Avg travel time for completed trips no matter the day category is around 50 mins. Most travelled ranges are from 40 to 65 mins.

````python
# Chart - 12 visualization code
# Heatmap between Distance and Day Category
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Convert timestamps
dataset['Request timestamp'] = pd.to_datetime(dataset['Request timestamp'])
dataset['Drop timestamp'] = pd.to_datetime(dataset['Drop timestamp'])

# Create Travel Time in minutes
dataset['Travel Time'] = (
    dataset['Drop timestamp'] - dataset['Request timestamp']
).dt.total_seconds() / 60

# Create Day Category
dataset['Day Category'] = dataset['Request timestamp'].dt.day_name()

# Average Travel Time by Day
heatmap_data = dataset.groupby('Day Category')['Travel Time'].mean().to_frame()

# Reorder days
days_order = ['Monday', 'Tuesday', 'Wednesday', 'Thursday',
              'Friday', 'Saturday', 'Sunday']

heatmap_data = heatmap_data.reindex(days_order)

# Plot Heatmap
plt.figure(figsize=(8,5))
sns.heatmap(
    heatmap_data,
    annot=True,
    fmt='.1f',
    cmap='YlOrRd'
)

plt.title('Heatmap of Average Travel Time by Day Category')
plt.ylabel('Day Category')
plt.xlabel('Travel Time')
plt.show()
````
